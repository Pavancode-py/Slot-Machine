```
# betting slot machine
import random
MAX_LINES=3
MAX_BET=100
MIN_BET=1

ROWS=3
COLUMNS=3

symbol_count={
    "A":2,
    "B":4,
    "C":6,
    "D":8
}

def slot_machine_spin(rows,columns,symbols):
    all_symbols=[]
    for symbol,symbol_count in symbols.items():
        for _ in range(symbol_count):
            all_symbols.append(symbol)

# ask user how much would like to deposit
def deposit():
    while True:
        amount=input("How much would you like to deposit?₹:")
        if amount.isdigit():
            amount=int(amount)
            if amount>0:
                break
            else:
                print("Amount must be greater than 0")
        else:
            print("Please enter a valid number")
    return amount

#  getting the lines

def getting_num_of_lines():
    while True:
        lines=input(f"Enter the number of lines to bet on (1 to {MAX_LINES}) ? :")
        if lines.isdigit():
            lines=int(lines)
            if 1 <=lines <= lines<=MAX_LINES:
                break
            else:
                print("Lines must be between 1 to 3 only")
        else:
            print("Please enter a valid number of lines")
    return lines

def get_bet():
    while True:
        amount=input("How much would you like to BET?₹:")
        if amount.isdigit():
            amount=int(amount)
            if 1<=amount<=MAX_BET:
                break
            else:
                print(f"Amount must be between ₹{MIN_BET} to ₹{MAX_BET}" )
        else:
            print("Please enter a valid number")
    return amount



def main():
    balance=deposit()
    lines=getting_num_of_lines()
    while True:
        bet=get_bet()
        total_bet=bet*lines
        if total_bet>balance:
            print(f"Insufficient balance. Your current balance is {balance}")
        else:
            break
    remaining_balance=balance-total_bet
    print(f"You are betting on ₹{bet} on {lines} lines. Total bet equals to ₹{total_bet}" )
    print(f"{remaining_balance} remaining balance")

main()




```
